```python
# drops the column which has all the rows empty

df = df.dropna(axis='columns', how = 'all')
df
```

```python
df.dtypes
```

```python
# Filling the missing values
# using mean

mean = np.mean(df.Cholesterol)
df.Cholesterol.fillna(mean, inplace=True)
print(df.Cholesterol)

df.isnull().any()
```

```python
# Filling the missing values
# using median

#median = np.median(df.Diabetes)
df.Diabetes.fillna(df.Diabetes.median(), inplace=True)
print(df.Diabetes)
df.isnull().any()
```

```python
# Renaming the columns

df = df.rename(columns={"Stress Level":"Stress_level"})
df
```

```python
# Filling the missing values
# using minimum value

min_value = df["Stress_level"].min()
print(min_value)

df.Stress_level.fillna(min_value, inplace=True)
print(df.Stress_level)
df.isnull().any()
```

```python
def box_plot(a):                       # Function to draw a box plot
    fig, ax = plt.subplots()           # Creating figure and axis

    ax.boxplot(a)                     # Plotting box plot for each column

    ax.set_xticklabels(['Age', 'Diabetes', 'Cholesterol'])      # labelling
    
    ax.set_title('Fitness')           # Set plot title and axis labels
    ax.set_xlabel('Factors')
    ax.set_ylabel('Output')
    plt.show()

Age = df.Age       # replacing data with our own dataset
Diabetes = df.Diabetes
Cholesterol= df.Cholesterol

box_plot([Age, Diabetes, Cholesterol])     #Function calling
```
