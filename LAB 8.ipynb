```python
df=df.head(15000)
```

```python
df=df.dropna(axis=0)
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn_extra.cluster import KMedoids

Clus_dataSet = StandardScaler().fit_transform(df)
clusterNum = 3
k_medoids = KMedoids(n_clusters=clusterNum, random_state=42)
k_medoids.fit(df)
labels = k_medoids.labels_
centroids = k_medoids.medoid_indices_
df["Clus_km"] = labels
print("\nPower Reading")
print(df["Clus_km"].value_counts())
print(df.head(10))
print("medoids are \n",centroids)
```
