```python
import pandas as pd
import numpy as np
data = pd.read_csv("Dataset_lab4.csv")
data
```

```python
data = {"Item_set" : [ "Milk,Bread,Biscuit","Bread,Milk,Biscuit,Cornflakes","Bread,Tea,Bournvita","Bread,Coffee,Bread,Milk",
    "Coffee,Tea,Biscuit","Bread,Tea,Bournvita","Coffee,Tea,Cornflakes","Coffee,Bread,Tea,Biscuit","Tea,Coffee,Bread,Coffee",
    "Bread,Milk","Coffee,Biscuit,Cornflakes","Bread,Biscuit,Cornflakes","Coffee,Sugar,Bournvita","Bread,Coffee,Butter",
    "Bread,Sugar,Biscuit","Coffee,Sugar,Cornflakes","Bread,Sugar,Bournvita","Bread,Coffee,Sugar","Bread,Coffee,Sugar,Butter",
    "Tea,Milk,Coffee,Cornflakes"]}

df = pd.DataFrame(data)

df['Item_set'] = df['Item_set'].apply(lambda x: x.split(','))    # Splitting the 'Item set' column into a list of items

for transaction in df['Item_set']:         # Printing the resulting list of dictionaries
    print(transaction)
```

```python
import mlxtend 
from mlxtend.preprocessing import TransactionEncoder

# Use TransactionEncoder
#TransactionEncoder is used for converting the transaction data into a suitable format for association rule mining.
#create an instance of the TransactionEncoder class, which will be used to convert the transaction data into a binary matrix format.

TE = TransactionEncoder()          

TE_Array = TE.fit(df['Item_set']).transform(df['Item_set'])

#TE_Array is a NumPy array that can be used as input for the Apriori algorithm
# df(item_set) represents a column in dataframe. Each row in this column contains a list of items purchased in a transaction. The fit method is used to fit the TransactionEncoder to the data, and the transform method converts the transaction data into a binary matrix.
# TE_Array can now be taken as input for the Apriori algorithm to discover frequent itemsets and mine association rules
```

```python
# Convert the result to a DataFrame
data = pd.DataFrame(TE_Array, columns=TE.columns_)

# Print the resulting DataFrame
print(data)

# Basically, converting the binary matrix obtained from the TransactionEncoder transformation into a pandas DataFrame.
#TE_Array is the binary matrix we got, 'columns' parameter is set to TE.columns_ so that unique itemnames is obtained during fit operation.
```

```python
# This imports the apriori function from the mlxtend.frequent_patterns module.
from mlxtend.frequent_patterns import apriori

frequent_itemsets=apriori(data,min_support=0.16,use_colnames=True)
frequent_itemsets

## Min_support sets the min threshold for an itemset to be considered frequent, here 16%
# use_colnames=True: This parameter indicates that the item names should be used in the resulting DataFrame 
```

```python
 from mlxtend.frequent_patterns import association_rules
 rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.16)
 rules
```
