```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
grades_df = pd.read_csv('C:/Users/ishit/Downloads/dataset.csv')
```

```python
grades_df.info()
```

```python
grades_df.describe()
```

```python
# a. Line Plot

plt.figure(figsize=(10, 5))
plt.plot(grades_df['Student'], grades_df['DBMS'], label='DBMS', marker='o')
plt.plot(grades_df['Student'], grades_df['Research_Methodology'], label='Research Methodology', marker='o')
plt.plot(grades_df['Student'], grades_df['DSA'], label='DSA', marker='o')
plt.title('Line Plot - Subject-wise Grades')
plt.xlabel('Student')
plt.ylabel('Grades')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# b. Bar Plot

grades_df.plot(kind='bar', x='Student', figsize=(10, 5))
plt.title('Bar Plot - Subject-wise Grades')
plt.xlabel('Student')
plt.ylabel('Grades')
plt.xticks(rotation=45)
plt.show()
```

```python
# c. Histogram

grades_df.plot(kind='hist', bins=20, alpha=0.5, figsize=(10, 5))
plt.title('Histogram - Subject-wise Grades')
plt.xlabel('Grades')
plt.ylabel('Frequency')
plt.show()
```

```python
# d. Scatter Plot

plt.figure(figsize=(8, 5))
sns.scatterplot(data=grades_df, x='DBMS', y='Research_Methodology', hue='DSA', s=100)
plt.title('Scatter Plot - DBMS vs Research Methodology with DSA as hue')
plt.xlabel('DBMS Grades')
plt.ylabel('Research Methodology Grades')
plt.show()
```

```python
# e. Pie Chart

subject_means = grades_df.mean()[['DBMS', 'Research_Methodology', 'DSA']]
plt.figure(figsize=(8, 8))
plt.pie(subject_means, labels=subject_means.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart - Average Grades')
plt.show()
```

```python
# f. Box Plot

grades_df.boxplot(column=['DBMS', 'Research_Methodology', 'DSA'], figsize=(10, 6))
plt.title('Box Plot - Subject-wise Grades')
plt.ylabel('Grades')
plt.show()
```
